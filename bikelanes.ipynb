```python
import numpy as np
import pandas as pd
import sklearn as sk
import seaborn as sns
import matplotlib as plt
```

```python
#Filtering for only household data

survey = pd.read_csv("attitudes2019.csv")
survey = survey[survey['weight'] != 0]

#Filtering households that refused to answer race data

survey = survey[survey['race_american_indian'] != 999]
survey = survey[survey['race_asian'] != 999]
survey = survey[survey['race_black'] != 999]
survey = survey[survey['race_native_hawaiian'] != 999]
survey = survey[survey['race_white'] != 999]
survey = survey[survey['race_other'] != 999]

#Filtering households who did not answer bike frequency question
survey = survey[survey['bike_freq'] != -9998]


survey
     
```

```python
race_bike = survey[['race_american_indian', 'race_asian',
                  'race_black', 'race_native_hawaiian', 'race_white', 'race_other', 'bike_freq']]

sns.heatmap(race_bike.corr())

race_bike.to_csv('racedata.csv')
```
